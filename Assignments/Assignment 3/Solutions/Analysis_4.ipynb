## Assignment 3_ Part 2_Question 2
- Use 'employee_compensation' data set.
- Find the people whose overtime salary is greater than 5% of Salaries (Salaries refers to 'Salaries' Column)
- For each job family these people are associated with (job family refers to 'Job Family' column), calculate what is the percentage of total benefits with respect to total compensation (So for each job family you have to calculate average total benefits and average total compensation). Create a new column (Percent_Total_Benefit) which has the percentage value. 
- **Display** the top 5 job families according to Percent_Total_Benefit **using df.head()**.  
- Write the output (Jobs and Percent_Total_Benefit) to a csv.

```python
import pandas as pd
```

```python
df = pd.read_csv("Data/employee_compensation.csv")
```

```python
df.head()
```

```python
df = df[df['Year Type']=='Calendar']
```

```python
# df.columns
```

```python
temp_df = round(df.groupby(['Job Family','Employee Identifier'],as_index=False)[['Salaries', 'Overtime', 'Other Salaries',
       'Total Salary', 'Retirement', 'Health/Dental', 'Other Benefits',
       'Total Benefits', 'Total Compensation']].mean(),2)
```

```python
temp_df.head()
```

```python
temp_df['Percent_Overtime'] = round(temp_df['Overtime']/temp_df['Salaries']*100,2)
```

```python
temp_df.head()
```

```python
overtime_df = temp_df[temp_df['Percent_Overtime'] > 5]
overtime_df.head()
```

```python
job_family_df = round(overtime_df.groupby(['Job Family'], as_index=False)['Total Benefits','Total Compensation'].mean(),2)
```

```python
job_family_df.head()
```

```python
job_family_df['Percent_Total_Benefit']=round((job_family_df['Total Benefits']/job_family_df['Total Compensation'])*100,2)
```

```python
job_family_df = job_family_df.sort_values(['Percent_Total_Benefit'],ascending=False).reset_index(drop=True)
```

```python
job_family_df.head()
```

```python
# Generate a csv file
job_family_df.to_csv('outputs/job_family.csv', index=False)
```
